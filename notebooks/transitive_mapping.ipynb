```python
import pandas as pd
import numpy as np
```

```python
df_wordNet_imageNet = pd.read_csv("./files/wordnet_to_imagenet.csv",sep=': ',names=["wordNet_class","imageNet_class"],header = None)
df_wordNet_imageNet
```

```python
df_wikidata_wordNet = pd.read_csv("./files/wikidata_to_wordnet.csv")
df_wikidata_wordNet.rename(columns={"item":"wikidata_class", "itemLabel":"item_label", "uri":"wordNet_class"})
```

```python
df_dbpedia_wikidata = pd.read_csv("./files/dbpedia_to_wikidata.csv")
df_dbpedia_wikidata= df_dbpedia_wikidata.rename(columns={"DBpediaClass":"dbpedia_class", "itemLabel":"item_label", "WikidataClass":"wikidata_class"})

df_dbpedia_wikidata[:20]
```

```python
df_dbpedia_wikidata
```

```python
s1 = pd.merge(df_wikidata_wordNet, df_dbpedia_wikidata, how='inner', on=['wikidata_class'])
s1.drop_duplicates("item_label", inplace = True) #we get duplicates because dbpedia_classes from another language are involved
s1
```

```python
#we make some preprocessing to get the uri in the correct format
uri_list = s1["uri"].to_list()

wordnet_list = []
for el in uri_list:
    foo = el.split("/")[4] #we split the string and receive the wordnet_class at foo[4]
    wd_class = foo.split("-")[0]
    wd_class = "n" + wd_class
    wordnet_list.append(wd_class)
wordnet_list
```

```python
wordnet_class = pd.DataFrame(wordnet_list, columns=["wordNet_class"])
baseline = pd.merge(wordnet_class, df_wordNet_imageNet, how='inner', on=['wordNet_class'])
baseline
```
