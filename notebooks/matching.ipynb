First we wrote some code to find an exact matching of the classes


```python
import logging

import pandas as pd
logger = logging.getLogger(__name__)
```

```python
class StringMatcher():    
    def __init__(self) -> None:
        self.df_dbpedia = pd.read_csv("./files/dbpedia_classes.csv")

        
    def match_exact_string(self, string:str) -> str:
        """method that gives us the dbpedia class to a given string
        Args: TODO:update description pls
            string (str): the class we are looking for
        Returns:
            pd.DataFrame: returns a DataFrame with the corresponding classes from dbpedia
        """
        substrings = string.split(",") 

        #we remove the blanks in front of every class
        stripped_substrings = [s.strip() for s in substrings]
        output = ""
        for str in stripped_substrings:
            if self.df_dbpedia['label'].eq(str).any():
                output = str      
            else:
                logger.info("Couldnt find matching string to: {str}")
        return output
    
    def match_lowercase_exact_string(self, string:str) -> str:
        substrings = string.split(",") 

        #we remove the blanks in front of every class
        stripped_substrings = [s.strip() for s in substrings]
        output = ""
        for str in stripped_substrings:
            if self.df_dbpedia['label'].eq(str.lower()).any():
                output = str      
            else:
                logger.info("Couldnt find matching string to: {str}")
        return output
    
      
    
```

We init our String Matcher, this class matches a String exactly to a dbpedia class and creates a dataframe that shows us the matching

```python
string_matcher = StringMatcher()

#dataframe with all imagenet classes
df_imgnet = pd.read_csv("./files/ImageNet.csv") 

#dataframe where we put in our results
df_exact_matching = pd.DataFrame(columns = ["image_net_class","dbpedia_class"])
list_exact_matching = []
for rowIndex, row in df_imgnet.iterrows():
    new_row = string_matcher.match_exact_string(row[1])
    if new_row:
        logger.info("ImageNet Class: {new_row}, DBpedia Class. {row[1]}")
        df_exact_matching.append({'image_net_class': new_row, "dbpedia_class": row[1]}, ignore_index=True)
        list_exact_matching.append([new_row, row[1]])

#df_exact_matching.head(100)
#print(list_exact_matching)
df = pd.DataFrame (list_exact_matching, columns = ['dbpedia_classes', 'image_net_classes'])
df.head(100)
```

As a next step we use another method of our StringMatcher class to find classes that doesnt have an exact matching but still match if we change to class from Image_net to a lowercase string

```python
string_matcher = StringMatcher()

#dataframe with all imagenet classes
df_imgnet = pd.read_csv("./files/ImageNet.csv") 

#dataframe where we put in our results
list_exact_matching = []
for rowIndex, row in df_imgnet.iterrows():
    new_row = string_matcher.match_lowercase_exact_string(row[1])
    if new_row:
        logger.info("ImageNet Class: {new_row}, DBpedia Class. {row[1]}")
        
        list_exact_matching.append([new_row, row[1]])
        
df = pd.DataFrame (list_exact_matching, columns = ['dbpedia_classes', 'image_net_classes'])
df.head(100)
```

As we can see we still have 23 classes so nothing changes if we convert all inputs to a lowercase string

Now we save these 23 classes in their own .csv file

```python
df.to_csv('exact_matching.csv', index=False)
```
